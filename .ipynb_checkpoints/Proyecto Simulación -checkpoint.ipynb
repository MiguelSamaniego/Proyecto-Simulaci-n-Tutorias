**Examen Interciclo: Simulación de Tutorías en la UPS**

**Objetivo:**

Desarrollar una comprensión práctica de los conceptos de simulación de eventos discretos mediante la creación de un modelo que simule las tutorías en la Universidad Politécnica Salesiana (UPS), enfocándose en la asignación de tutores y la gestión de aulas.

**Enunciado:**

Se requiere diseñar y desarrollar un programa o script para simular el sistema de tutorías en la UPS. El proceso se describe de la siguiente manera:

En la UPS, los estudiantes (tutorados) solicitan tutorías en diversas materias: Cálculo Diferencial, Programación, Química y Álgebra Lineal. Las probabilidades de que soliciten tutorías en estas materias son del 40%, 25%, 10% y 25% respectivamente. Se observa una mayor demanda de tutorías en las tardes comparado con las mañanas.

La disponibilidad de tutores es la siguiente: 10 para Cálculo, 10 para Programación, 2 para Química y 3 para Álgebra Lineal. Una vez que un tutorado solicita ayuda. Si un tutor está disponible, deben esperar 10 minutos antes de poder solicitar una de las 20 aulas disponibles.

La duración promedio de las tutorías para cada asignatura es:

- Cálculo: 150 minutos ± 30 minutos
- Química: 120 minutos ± 15 minutos
- Programación: 120 minutos ± 20 minutos
- Álgebra Lineal: 60 minutos ± 10 minutos

Si un tutorado debe esperar a que un tutor se libere, el tiempo de espera antes de retirarse por falta de tutores es de 15 a 20 minutos.

**Requerimientos del Programa:**

1. El programa debe simular al menos 500 unidades de tiempo.
2. Debe poder generar gráficos que muestren:
   - La cantidad de tutorías solicitadas y completadas por materia.
   - La distribución del tiempo de espera antes de ser atendidos.
   - La utilización de las aulas.
3. Calcular el tiempo promedio desde la solicitud hasta la finalización de la tutoría, por materia.
4. Implementar y analizar cualquier otra fase que se considere relevante para el estudio.

**Salida Esperada:**

Proporcione estadísticas detalladas y gráficos que muestren el rendimiento del sistema de tutorías, incluyendo:

- Tiempo promedio de espera por tutoría, por materia.
- Porcentaje de tutorías completadas.
- Tiempo promedio de uso de aulas.

**Conclusión:**

Analice los datos obtenidos para ofrecer sugerencias que puedan mejorar la eficiencia del sistema de tutorías en la UPS.

---

Este enunciado proporciona una estructura clara para un proyecto de simulación, estableciendo objetivos, describiendo el proceso a simular, definiendo los requerimientos del programa y la salida esperada, y pidiendo una conclusión basada en el análisis de los datos simulados.

```python
import simpy
import random
import matplotlib.pyplot as plt
```

```python


# Parámetros de la simulación
TIEMPO_SIMULACION = 60*8
NUM_TUTORES = {'Cálculo': 10, 'Programación': 10, 'Química': 2, 'Álgebra Lineal': 3}
NUM_AULAS = 20
TIEMPOS_TUTORIAS = {'Cálculo': (150, 30), 'Programación': (120, 20), 'Química': (120, 15), 'Álgebra Lineal': (60, 10)}
PROBABILIDAD_DISPONIBILIDAD = 0.8
TIEMPO_ESPERA_MAXIMO = 20

class Estudiante:
    def __init__(self, env, tutoria, materia):
        self.env = env
        self.tutoria = tutoria
        self.materia = materia
        self.tiempo_espera = 0
        self.env.process(self.solicitar_tutoria())

    def solicitar_tutoria(self):
        with self.tutoria.tutores[self.materia].request() as req:
            inicio_espera = self.env.now
            resultado = yield req | self.env.timeout(random.randint(15, TIEMPO_ESPERA_MAXIMO))

            if req in resultado:
                self.tiempo_espera = self.env.now - inicio_espera
                tiempo_tutoria = random.gauss(*TIEMPOS_TUTORIAS[self.materia])
                yield self.env.timeout(10)  # Espera para el aula
                with self.tutoria.aulas.request() as aula_req:
                    yield aula_req
                    yield self.env.timeout(tiempo_tutoria)
                self.tutoria.estadisticas['completadas'] += 1
                self.tutoria.estadisticas['uso_aulas'].append(tiempo_tutoria)
            else:
                self.tiempo_espera = self.env.now - inicio_espera
                self.tutoria.estadisticas['tiempo_espera'].append(self.tiempo_espera)

class Tutoria:
    def __init__(self, env, num_tutores, num_aulas):
        self.env = env
        self.tutores = {materia: simpy.Resource(env, num_tutores[materia]) for materia in num_tutores}
        self.aulas = simpy.Resource(env, num_aulas)
        self.estadisticas = {'solicitadas': 0, 'completadas': 0, 'tiempo_espera': [], 'uso_aulas': []}
        self.estadisticas_por_materia = {materia: {'solicitadas': 0, 'completadas': 0} for materia in NUM_TUTORES.keys()}

    def actualizar_estadisticas(self, materia, completada):
        self.estadisticas_por_materia[materia]['solicitadas'] += 1
        if completada:
            self.estadisticas_por_materia[materia]['completadas'] += 1


def generar_solicitud(env, tutoria):
    materias = ['Cálculo', 'Programación', 'Química', 'Álgebra Lineal']
    probabilidad = [0.4, 0.25, 0.1, 0.25]
    while True:
        materia = random.choices(materias, probabilidad)[0]
        Estudiante(env, tutoria, materia)
        tutoria.estadisticas['solicitadas'] += 1
        yield env.timeout(random.expovariate(1/30))

env = simpy.Environment()
tutoria = Tutoria(env, NUM_TUTORES, NUM_AULAS)
env.process(generar_solicitud(env, tutoria))
env.run(until=TIEMPO_SIMULACION)

# Análisis de resultados y generación de gráficos
# ...
```

```python
# Análisis y visualización de resultados
print(f"Tutorías Solicitadas: {tutoria.estadisticas['solicitadas']}")
print(f"Tutorías Completadas: {tutoria.estadisticas['completadas']}")
print(f"Porcentaje de Tutorías Completadas: {100 * tutoria.estadisticas['completadas'] / tutoria.estadisticas['solicitadas']:.2f}%")

if tutoria.estadisticas['tiempo_espera']:
    print(f"Tiempo de espera promedio: {sum(tutoria.estadisticas['tiempo_espera']) / len(tutoria.estadisticas['tiempo_espera']):.2f} minutos")

if tutoria.estadisticas['uso_aulas']:
    print(f"Uso promedio de aulas: {sum(tutoria.estadisticas['uso_aulas']) / len(tutoria.estadisticas['uso_aulas']):.2f} minutos")

# Generación de gráficos
plt.hist(tutoria.estadisticas['tiempo_espera'], bins=10, color='blue', edgecolor='black')
plt.title("Distribución del Tiempo de Espera de los Estudiantes")
plt.xlabel("Tiempo de Espera (minutos)")
plt.ylabel("Número de Estudiantes")
plt.show()
```

```python
materias = NUM_TUTORES.keys()
solicitadas = [tutoria.estadisticas_por_materia[materia]['solicitadas'] for materia in materias]
completadas = [tutoria.estadisticas_por_materia[materia]['completadas'] for materia in materias]

x = range(len(materias))
plt.bar(x, solicitadas, width=0.4, label='Solicitadas', align='center')
plt.bar(x, completadas, width=0.4, label='Completadas', align='edge')
plt.xlabel('Materias')
plt.ylabel('Número de Tutorías')
plt.title('Tutorías Solicitadas y Completadas por Materia')
plt.xticks(x, materias)
plt.legend()
plt.show()

# Gráfico del uso de aulas
plt.hist(tutoria.estadisticas['uso_aulas'], bins=10, color='green', edgecolor='black')
plt.title("Uso de Aulas")
plt.xlabel("Duración (minutos)")
plt.ylabel("Frecuencia")
plt.show()
```
